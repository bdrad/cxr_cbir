```python
import pandas as pd
import numpy as np

metadata_path = '../../ChestX-ray14/Data_Entry_2017_v2020.csv'
metadata = pd.read_csv(metadata_path)

class_mappings = {
    'Cardiomegaly': 'Cardiomegaly',
    'Opacity': 'Atelectasis|Efussion|Infiltration|Pneumonia|Consolidation|Edema|Fibrosis|Pleural_Thickening',
    'Emphysema': 'Emphysema'
}

retrieval_datasets_metadata = pd.DataFrame(columns=['Task', 'Number of Subjects', 'Male', 'Female', 'Positive', 'Negative'])

data_dict = {
    
}

for task in class_mappings:
    data_dict['Task'] = task
    positive_metadata = metadata[metadata['Finding Labels'].str.contains(class_mappings[task])]
    positive_metadata = positive_metadata.groupby(['Patient ID', 'Patient Gender']).size().reset_index(name='Freq')
    female_pos_count = (positive_metadata['Patient Gender'].value_counts()['F'])
    male_pos_count = (positive_metadata['Patient Gender'].value_counts()['M'])
    pos_count = female_pos_count + male_pos_count
    
    negative_metadata = metadata[metadata['Finding Labels'].str.contains('No Finding')][:pos_count]
    negative_metadata = negative_metadata.groupby(['Patient ID', 'Patient Gender']).size().reset_index(name='Freq')
    female_neg_count = (negative_metadata['Patient Gender'].value_counts()['F'])
    male_neg_count = (negative_metadata['Patient Gender'].value_counts()['M'])
    neg_count = female_neg_count + male_neg_count
    
    female_count = female_pos_count + female_neg_count
    male_count = male_pos_count + male_neg_count
    
    total_count = male_count + female_count
        

    retrieval_datasets_metadata.loc[len(retrieval_datasets_metadata)] = {
        'Task': task,
        'Number of Subjects': total_count,
        'Male': male_count,
        'Female': female_count,
        'Positive': pos_count,
        'Negative': neg_count
    }
    

retrieval_datasets_metadata
```
