```python
from dataset import ChestXray14
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import torch
```

```python
def show_first_img(dataset):
    train_data_loader = DataLoader(dataset, shuffle=False)
    it = iter(train_data_loader)
    im = next(it)[0]
    im = im.squeeze(0)
    plt.imshow(im.permute(1, 2, 0))
    plt.show()
```

```python
class_name = 'Cardiomegaly'
train_dataset_orig = ChestXray14(phase='train', class_name=class_name, transform_type='orig')
train_dataset_norm = ChestXray14(phase='train', class_name=class_name, transform_type='norm')
train_dataset_HE = ChestXray14(phase='train', class_name=class_name, transform_type='HE')
train_dataset_CLAHE = ChestXray14(phase='train', class_name=class_name, transform_type='CLAHE')
train_dataset_complement = ChestXray14(phase='train', class_name=class_name, transform_type='complement')
train_dataset_gamma_correct = ChestXray14(phase='train', class_name=class_name, transform_type='gamma_correct')
train_dataset_BCET = ChestXray14(phase='train', class_name=class_name, transform_type='BCET')
```

```python
show_first_img(train_dataset_orig)
```

```python
show_first_img(train_dataset_norm)
```
